```python
import os
import json
import requests
```

```python
folder_path = 'sample_traces/'
```

```python
file_names = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
```

```python
print(file_names)
```

example: https://samples.adsbexchange.com/traces/2017/04/01/03/trace_full_00b203.json

```python
url_list = []
for val in file_names:
    year = val[0:4]
    if year in ['2022', '2023', '2024']:
        month = val[5:7]
        last_two_digits = val[12:14]
        plane = val[8:14]
        url_string = f'https://samples.adsbexchange.com/hires-traces/{year}/{month}/01/{last_two_digits}/trace_full_{plane}.json'
        url_list.append(url_string)
    else:
        pass

```

```python
for url in url_list:
    print(url)
    file_name = f'{url[46:50]}_{url[51:53]}_{url[54:56]}_{url[-11:-5]}.json'
    print(file_name)
```

```python
print(len(url_list))
```

```python
counter= 0
for url in url_list:
    try:
        print(counter)
        print(f'getting info from {url}')
        counter += 1
        
        output_folder = 'sample_highres_traces/'
        response = requests.get(url)
        response.raise_for_status() 
        #change the file name to include the day date - it will be always 01 but for the sake of clairy
        file_name = f'{url[46:50]}_{url[51:53]}_{url[54:56]}_{url[-11:-5]}.json'

        json_data = response.json() 
        
        file_name = os.path.join(output_folder, file_name)
        with open(file_name, 'w') as json_file:
            json.dump(json_data, json_file, indent=2)
            print(f"Data from {url} saved to {file_name}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from {url}: {e}")

    except json.JSONDecodeError as e:
        print(f"Error parsing JSON from {url}: {e}")
```
