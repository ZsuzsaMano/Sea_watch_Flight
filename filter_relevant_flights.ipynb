### Check the number of points in traces array

```python
import os
import json

directory_in_str = "sample_traces"
path = "./sample_traces"

irrelevants_path = "./too_short"
os.makedirs(irrelevants_path, exist_ok=True)

dirs = os.listdir(path)

for file in dirs:
    file_path = os.path.join(path, file)  # Construct the full path to the file
    with open(file_path, 'r') as f:
        data = json.load(f)  # Assuming the file content is in JSON format
        traces_array = data.get('trace', [])  # Assuming 'traces' is the key for the array
        traces_array_length = len(traces_array)
        if traces_array_length > 45:  # Check if traces array length is greater than 45
            print("File:", file, ", Traces array length:", traces_array_length)
        else:
            print("Moving file:", file)
            irrelevants_file_path = os.path.join(irrelevants_path, file)  # Destination file path
            os.replace(file_path, irrelevants_file_path)  # Move the file to the new directory

```
